```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(df10)
df11 = scaler.transform(df10)
```

```python
df11
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df11)
```

```python
y = kmeans.predict(df11)
```

```python
y
```

```python
df10['cluster'] = y
```

```python
df10.head()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.boxplot(x="cluster", y="MEDIA_9EF_LP", hue="SOCIO_ECON",  data=df10, ax=ax1)

sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.boxplot(x="cluster", y="MEDIA_9EF_MT", hue="SOCIO_ECON", data=df10, ax=ax2)
```
